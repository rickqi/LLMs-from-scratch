## 练习 5.1：温度缩放的 softmax 分数与采样概率

**练习目的**：理解温度（temperature）如何影响 softmax 概率分布和采样结果。

**关键概念**：
- 温度越低（如 0.1），softmax 越尖锐，近乎贪心搜索（只选最佳 token）
- 温度越高（如 5），softmax 越平坦，低概率 token 也有机会被选中
- 温度=1 时为原始 softmax，不做缩放

- 我们可以使用本章定义的 `print_sampled_tokens` 函数来统计 "pizza" 这个词被采样到的次数
- 从第 5.3.1 节的代码开始

- 当温度为 0 或 0.1 时，"pizza" 被采样 0 次；当温度升高到 5 时，被采样 32 次。估计概率 = 32/1000 x 100% = 3.2%

- 实际概率为 4.3%，包含在缩放后的 softmax 概率张量中（`scaled_probas[2][6]`）

- 以下是使用第 5 章代码的完整独立示例：

````python
import torch

vocab = { 
    "closer": 0,
    "every": 1, 
    "effort": 2, 
    "forward": 3,
    "inches": 4,
    "moves": 5, 
    "pizza": 6,
    "toward": 7,
    "you": 8,
} 
inverse_vocab = {v: k for k, v in vocab.items()}

next_token_logits = torch.tensor(
    [4.51, 0.89, -1.90, 6.75, 1.63, -1.62, -1.89, 6.28, 1.79]
)

def print_sampled_tokens(probas):
    torch.manual_seed(123)
    sample = [torch.multinomial(probas, num_samples=1).item() for i in range(1_000)]
    sampled_ids = torch.bincount(torch.tensor(sample))
    for i, freq in enumerate(sampled_ids):
        print(f"{freq} x {inverse_vocab[i]}")


def softmax_with_temperature(logits, temperature):
    scaled_logits = logits / temperature
    return torch.softmax(scaled_logits, dim=0)


temperatures = [1, 0.1, 5]  # Original, higher, and lower temperature
scaled_probas = [softmax_with_temperature(next_token_logits, T) for T in temperatures]
````

- 现在，我们可以遍历 `scaled_probas`，打印每种温度下的采样频率：

````python
for i, probas in enumerate(scaled_probas):
    print("\n\nTemperature:", temperatures[i])
    print_sampled_tokens(probas)
````

- 注意，采样是对实际概率的近似估计
- 例如，如果 "pizza" 被采样 32/1000 次，估计概率为 3.2%
- 要获取实际概率，可以直接查看 `scaled_probas` 中对应的条目

- "pizza" 是词表中的第 7 个条目（索引 6），对于温度 5，我们可以这样获取：

````python
temp5_idx = 2
pizza_idx = 6

scaled_probas[temp5_idx][pizza_idx]
````

在温度设为 5 时，"pizza" 被采样到的实际概率为 4.3%

```
温度对比总结:
  T=0.1: pizza 采样 0 次  → 太低，几乎只选 forward
  T=1  : pizza 采样 0 次  → 默认概率仍太低
  T=5  : pizza 采样 32 次 → 高温度放大了低概率词的采样机会
```

**练习目的**：理解温度和 top-k 的调参策略。

- 温度和 top-k 设置需要根据具体的 LLM 进行调整（一种试错过程，直到生成理想的输出）

## 练习 5.3：解码函数的确定性行为

**练习目的**：找到让 `generate` 函数产生确定性（可复现）输出的方法。

有多种方法可以强制 `generate` 函数产生确定性行为：

1. 设置 `temperature=0.0`（缩放 logits 后 softmax 变为 argmax）
2. 设置 `top_k=1`（只从最优示的 1 个 token 中选择）

```
两种方法等价:
  temperature=0 → logits/0 → 最大 logit → +inf → softmax=1.0
  top_k=1      → 只保留最大 logit 的 token → 必然被选中
  → 两者都等价于 argmax（贪心解码）
```

以下是使用第 5 章代码的完整独立示例：

````python
import tiktoken
import torch
from previous_chapters import GPTModel


GPT_CONFIG_124M = {
    "vocab_size": 50257,  # Vocabulary size
    "context_length": 256,       # Shortened context length (orig: 1024)
    "emb_dim": 768,       # Embedding dimension
    "n_heads": 12,        # Number of attention heads
    "n_layers": 12,       # Number of layers
    "drop_rate": 0.1,     # Dropout rate
    "qkv_bias": False     # Query-key-value bias
}


torch.manual_seed(123)

tokenizer = tiktoken.get_encoding("gpt2")
model = GPTModel(GPT_CONFIG_124M)
model.load_state_dict(torch.load("model.pth", weights_only=True))
model.eval();
````

````python
from gpt_generate import generate, text_to_token_ids, token_ids_to_text
from previous_chapters import generate_text_simple
````

````python
# Deterministic function that used torch.argmax

start_context = "Every effort moves you"

token_ids = generate_text_simple(
    model=model,
    idx=text_to_token_ids(start_context, tokenizer),
    max_new_tokens=25,
    context_size=GPT_CONFIG_124M["context_length"]
)

print("Output text:\n", token_ids_to_text(token_ids, tokenizer))
````

````python
# Deterministic behavior: No top_k, no temperature scaling

token_ids = generate(
    model=model,
    idx=text_to_token_ids("Every effort moves you", tokenizer),
    max_new_tokens=25,
    context_size=GPT_CONFIG_124M["context_length"],
    top_k=None,
    temperature=0.0
)

print("Output text:\n", token_ids_to_text(token_ids, tokenizer))
````

````python
import tiktoken
import torch
from previous_chapters import GPTModel


GPT_CONFIG_124M = {
    "vocab_size": 50257,   # Vocabulary size
    "context_length": 256, # Shortened context length (orig: 1024)
    "emb_dim": 768,        # Embedding dimension
    "n_heads": 12,         # Number of attention heads
    "n_layers": 12,        # Number of layers
    "drop_rate": 0.1,      # Dropout rate
    "qkv_bias": False      # Query-key-value bias
}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

tokenizer = tiktoken.get_encoding("gpt2")

checkpoint = torch.load("model_and_optimizer.pth", weights_only=True)
model = GPTModel(GPT_CONFIG_124M)
model.load_state_dict(checkpoint["model_state_dict"])
model.to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=0.0004, weight_decay=0.1)
optimizer.load_state_dict(checkpoint["optimizer_state_dict"])
model.train();
````

## 练习 5.5：预训练模型的训练和验证集损失

**练习目的**：计算 OpenAI 预训练权重加载后的 GPT-2 在 The Verdict 数据集上的训练/验证损失。

- 我们可以使用以下代码计算 GPT 模型的训练和验证集损失：

```python
train_loss = calc_loss_loader(train_loader, gpt, device)
val_loss = calc_loss_loader(val_loader, gpt, device)
```

- 124M 参数模型的结果如下：

```
Training loss: 3.754748503367106
Validation loss: 3.559617757797241
```

- 主要观察：训练和验证集的性能在同一个量级，这可能有多种解释：

1. The Verdict 不在 OpenAI 训练 GPT-2 的预训练数据集中。因此模型没有对训练集明显过拟合，在训练和验证集上表现相似。（验证损失略低于训练损失，这在深度学习中不寻常，但可能是因为数据集较小的随机噪声。）

2. The Verdict 是 GPT-2 训练数据集的一部分。这种情况下我们无法判断模型是否过拟合，因为验证集可能也被用于了训练。要评估过拟合程度，需要一个 OpenAI 完成 GPT-2 训练后生成的新数据集。

以下代码是为此新 notebook 复现的完整独立示例。

````python
import tiktoken
import torch
from previous_chapters import GPTModel


GPT_CONFIG_124M = {
    "vocab_size": 50257,   # Vocabulary size
    "context_length": 256, # Shortened context length (orig: 1024)
    "emb_dim": 768,        # Embedding dimension
    "n_heads": 12,         # Number of attention heads
    "n_layers": 12,        # Number of layers
    "drop_rate": 0.1,      # Dropout rate
    "qkv_bias": False      # Query-key-value bias
}


torch.manual_seed(123)

tokenizer = tiktoken.get_encoding("gpt2")
````

````python
from gpt_download import download_and_load_gpt2

settings, params = download_and_load_gpt2(model_size="124M", models_dir="gpt2")
````

````python
# Define model configurations in a dictionary for compactness
model_configs = {
    "gpt2-small (124M)": {"emb_dim": 768, "n_layers": 12, "n_heads": 12},
    "gpt2-medium (355M)": {"emb_dim": 1024, "n_layers": 24, "n_heads": 16},
    "gpt2-large (774M)": {"emb_dim": 1280, "n_layers": 36, "n_heads": 20},
    "gpt2-xl (1558M)": {"emb_dim": 1600, "n_layers": 48, "n_heads": 25},
}

# Copy the base configuration and update with specific model settings
model_name = "gpt2-small (124M)"  # Example model name
NEW_CONFIG = GPT_CONFIG_124M.copy()
NEW_CONFIG.update(model_configs[model_name])
NEW_CONFIG.update({"context_length": 1024, "qkv_bias": True})

gpt = GPTModel(NEW_CONFIG)
gpt.eval();
````

````python
from gpt_generate import load_weights_into_gpt


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
load_weights_into_gpt(gpt, params)
gpt.to(device);
````

````python
import os
import urllib.request
from previous_chapters import create_dataloader_v1


file_path = "the-verdict.txt"
url = "https://raw.githubusercontent.com/rasbt/LLMs-from-scratch/main/ch02/01_main-chapter-code/the-verdict.txt"

if not os.path.exists(file_path):
    with urllib.request.urlopen(url) as response:
        text_data = response.read().decode('utf-8')
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(text_data)
else:
    with open(file_path, "r", encoding="utf-8") as file:
        text_data = file.read()


# Train/validation ratio
train_ratio = 0.90
split_idx = int(train_ratio * len(text_data))
train_data = text_data[:split_idx]
val_data = text_data[split_idx:]


torch.manual_seed(123)

train_loader = create_dataloader_v1(
    train_data,
    batch_size=2,
    max_length=GPT_CONFIG_124M["context_length"],
    stride=GPT_CONFIG_124M["context_length"],
    drop_last=True,
    shuffle=True,
    num_workers=0
)

val_loader = create_dataloader_v1(
    val_data,
    batch_size=2,
    max_length=GPT_CONFIG_124M["context_length"],
    stride=GPT_CONFIG_124M["context_length"],
    drop_last=False,
    shuffle=False,
    num_workers=0
)
````

````python
from gpt_train import calc_loss_loader

torch.manual_seed(123) # For reproducibility due to the shuffling in the data loader
train_loss = calc_loss_loader(train_loader, gpt, device)
val_loss = calc_loss_loader(val_loader, gpt, device)

print("Training loss:", train_loss)
print("Validation loss:", val_loss)
````

```
模型对比：
  GPT-2 small (124M): train=3.75, val=3.56
  GPT-2 XL   (1558M): train=3.30, val=3.12  ← 更大模型损失更低
  → 更大的模型对语言理解更好
```

````python
settings, params = download_and_load_gpt2(model_size="1558M", models_dir="gpt2")

model_name = "gpt2-xl (1558M)"
NEW_CONFIG = GPT_CONFIG_124M.copy()
NEW_CONFIG.update(model_configs[model_name])
NEW_CONFIG.update({"context_length": 1024, "qkv_bias": True})

gpt = GPTModel(NEW_CONFIG)
gpt.eval()

load_weights_into_gpt(gpt, params)
gpt.to(device)

torch.manual_seed(123)
train_loss = calc_loss_loader(train_loader, gpt, device)
val_loss = calc_loss_loader(val_loader, gpt, device)

print("Training loss:", train_loss)
print("Validation loss:", val_loss)
````
